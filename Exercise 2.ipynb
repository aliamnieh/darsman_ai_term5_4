````python
# TODO : رسم هیستوگرام قیمت

plt.figure(figsize=(15,8))
counts,bin_edges,patches=plt.hist(df['Price ($)'],bins=120,edgecolor='black',alpha=0.7,color='green')

bin_centers=(bin_edges[:-1]+bin_edges[1:])/2
plt.plot(bin_centers,counts,linestyle='-',color='red',label='curve')

plt.title('Histogram for Price')
plt.xlabel('price range')
plt.ylabel('Quantity')
plt.legend()
plt.show()
````

````python
# TODO: جعبه‌ای قیمت لپ‌تاپ‌های هر برند
import math
brands=df['Brand'].value_counts().index
plt.figure(figsize=(20,50))

for i,brand in enumerate(brands):
    plt.subplot(math.ceil(len(brands)/2),2,i+1)
    plt.boxplot(df[df['Brand']==brand]['Price ($)'].values,labels=[brand])
    plt.title(brand)

plt.tight_layout()
plt.show()
````

````python
import math
brands=df['Brand'].value_counts().index
plt.figure(figsize=(30,50))

for i,brand in enumerate(brands):
    plt.subplot(math.ceil(len(brands)/2),2,i+1)
    df[df['Brand']==brand]['Price ($)'].hist(bins=100)
    plt.title(brand)

plt.tight_layout()
plt.show()
````

````python
# TODO: متدی برای پیش پردازش  فضای ذخیره سازی طبق توضیحات ارائه شده بنویسید.
def parse_storage(storage_str):
    match=re.search(r'\d+',storage_str)
    if match:
        number=int(match.group())
        if 'GB' in storage_str:
            if 'SSD' in storage_str:
                return (number,'SSD')
            elif 'HDD' in storage_str:
                return (number,'HDD')
        elif 'TB' in storage_str:
            if 'SSD' in storage_str:
                return (number * 1024,'SSD')
            elif 'HDD' in storage_str:
                return (number*1024,'HDD')
            
df['Storage']=df['Storage'].apply(parse_storage)

df['Storage_Capacity_GB']=df['Storage'].apply(lambda x:x[0])
df['Storage_Type']=df['Storage'].apply(lambda x:x[1])
df.head(50)
    





````

````python
# TODO متدی برای مدیریت ویژگی های GPU و Processor بنویسید.

def parse_processor(str1):
    if 'Intel' in str1:
        if '3' in str1:
            return ('Intel',3)
        elif '5' in str1:
            return ('Intel',5)
        elif '7' in str1:
            return ('Intel',7)
        elif '9' in str1:
            return ('Intel',9)
    elif 'AMD' in str1:
        if '3' in str1:
            return ('AMD',3)
        elif '5' in str1:
            return ('AMD',5)
        elif '7' in str1:
            return ('AMD',7)
        elif '9' in str1:
            return ('AMD',9)
        
df['Processor']=df['Processor'].apply(parse_processor)
df['Processor_Brand']=df['Processor'].apply(lambda x:x[0])
df['Processor_Core']=df['Processor'].apply(lambda x:x[1])
df.head(20)
````

````python
# TODO: متد منحنی یادگیری را بنویسید
def plot_learning_curve(model, X, y, model_name):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator=model,
        X=X,
        y=y,
        cv=5,
        scoring="neg_mean_squared_error",
        train_sizes = np.linspace(0.1,1.0,10)
    )

    train_mean = -train_scores.mean(axis=1)
    test_mean = -test_scores.mean(axis=1)

    plt.subplots(figsize=(10,8))
    plt.plot(train_sizes, train_mean, label="train")
    plt.plot(train_sizes, test_mean, label="validation")

    plt.title(f"Learning Curve for {model_name}")
    plt.xlabel("Training Set Size")
    plt.ylabel("MSE")
    plt.legend(loc="best")

    plt.show()
````

````python
# TODO: تابع برای محاسبه و نمایش معیارهای ارزیابی
def evaluate_model(pipeline, X, y, dataset_name):
    y_pred=pipeline.predict(X)
    mse=mean_squared_error(y,y_pred)
    rmse=root_mean_squared_error(y,y_pred)
    mae=mean_absolute_error(y,y_pred)
    r2=r2_score(y,y_pred)
    print(f'Model Evaluation On Sets: {dataset_name}')
    print(f'Mean Squared Error (MSE): {mse}')
    print(f'Root Mean Squared Error (RMSE): {rmse}')
    print(f'Mean Absolute Error (MAE): {mae}')
    print(f'R-squared (R2): {r2}')
    return {'MSE':round(mse,3),'RMSE':round(rmse,3),'MAE':round(mae,3),'R2':round(r2,3)}


````

````python
# TODO: نمایش نمودار پراکندگی بین مقادیر واقعی و پیش‌بینی شده
def plot_prediction_scatter(y_train,y_train_pred,y_test,y_test_pred,y_validation,y_validation_pred):
    plt.figure(figsize=(20,25))

    plt.subplot(2,2,1)
    plt.scatter(y_train,y_train_pred,color='Red')
    plt.plot(y_train,y_train,linestyle='-',color='Blue')
    plt.xlabel('y_train')
    plt.ylabel('y_train_pred')
    plt.title('Train')

    plt.subplot(2,2,2)
    plt.scatter(y_test,y_test_pred,color='Red')
    plt.plot(y_test,y_test,linestyle='-',color='Blue')
    plt.xlabel('y_test')
    plt.ylabel('y_test_pred')
    plt.title('Test')

    plt.subplot(2,2,3)
    plt.scatter(y_validation,y_validation_pred,color='Red')
    plt.plot(y_validation,y_validation,linestyle='-',color='Blue')
    plt.xlabel('y_validation')
    plt.ylabel('y_validation_pred')
    plt.title('Validation')

    plt.subplot(2,2,4)
    plt.scatter(y_test,(y_test-y_test_pred),linestyle='-',color='Blue')
    plt.xlabel('y_test')
    plt.ylabel('Difference')
    plt.title('Difference ')

    plt.tight_layout()
    plt.show()

# این تابع را به گونه ای نوشتیم که موقعیت نقاط را نسبت به یک خط نشان دهد
# فاصله از این خط نشان میدهد که مقدار پیش بینی شده با مقدار در واقعیت چقدر تفاوت دارند
````
